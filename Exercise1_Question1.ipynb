### Dataset - chickwts.txt

The dataset describes attributes of chicks. The first column contatins "weights", which is a numeric variable. The second column is "feed", which is a character string describing the type of feed given to the chick.


```python
import pandas 
chData=pandas.read_csv('chickwts.txt')
chData.head()
```

```python
from plotnine import *
import numpy
plot1=ggplot(chData, aes(y="weight", x="feed"))+geom_bar(stat= "summary", fun_y=numpy.mean)
plot1+theme_bw()+xlab("Feed Type")+ylab("Average Weight of Chicks")
```
